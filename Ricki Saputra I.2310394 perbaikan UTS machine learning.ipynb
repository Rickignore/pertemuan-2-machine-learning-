```python
import os
import pandas as pd

dataset_path = '/kaggle/input/e-commerce-customer-for-behavior-analysis'
file = 'ecommerce_customer_data_large.csv'
data = pd.read_csv(os.path.join(dataset_path, file))
df = pd.DataFrame(data)
df
```

#Linear regresion

```python
X = data[['Product Price', 'Quantity', 'Customer Age', 'Churn']]
y = data[['Total Purchase Amount']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Linear Regression
model = LinearRegression()

# Training
model.fit(X_train, y_train)

# Prediksi
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

# Evaluasi Training
mse_training = mean_squared_error(y_train, y_pred_train)
print('Hasil Evaluasi MSE Training =', mse_training)

# Evaluasi Testing
mse_testing = mean_squared_error(y_test, y_pred_test)
print('Hasil Evaluasi MSE Testing =', mse_testing)
```

```python
import pandas as pd

coeff = model.coef_[0]
feature_importance = pd.DataFrame({'Feature': X.columns, 'Coefficient': coeff})
feature_importance = feature_importance.sort_values(by='Coefficient', ascending=False)
print(feature_importance)
```

### Model regresi di atas menghasilkan MSE training dan testing sekitar dua juta, yang berarti model cukup stabil tetapi masih belum akurat dalam memprediksi total pembelian.

### Analisis feature importance dari koefisien linear regression menunjukkan fitur mana yang paling mendorong nilai pembelian naik atau turun, di mana koefisien positif berarti fitur tersebut meningkatkan total pembelian dan koefisien negatif berarti menurunkannya.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

data_enc = pd.get_dummies(data, columns=['Gender'], drop_first=True)

X = data_enc[['Product Price', 'Quantity', 'Customer Age', 'Churn', 'Gender_Male']]
y = data_enc[['Total Purchase Amount']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

model = LinearRegression()
model.fit(X_train, y_train)

mean_age = data['Customer Age'].mean()
d = data_enc[data_enc['Customer Age'] > mean_age]
pred = model.predict(d[['Product Price','Quantity','Customer Age','Churn','Gender_Male']])

male_avg = pred[d['Gender_Male']==1].mean()
female_avg = pred[d['Gender_Male']==0].mean()

print("Male:", male_avg)
print("Female:", female_avg)
```

### Pembeli laki-laki tidak membeli lebih banyak daripada perempuan pada usia di atas rata-rata; rata-rata pembelian keduanya hampir sama, berdasarkan hasil di atas Female lebih besar sekitar 0.5–0.6 poin.

```python
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

X_sample, y_sample = X.sample(500, random_state=1), y.sample(500, random_state=1)
X_train, X_test, y_train, y_test = train_test_split(X_sample, y_sample, test_size=0.2, random_state=1)

models = {
    'Linear': LinearRegression(),
    'Optimal': make_pipeline(PolynomialFeatures(5), LinearRegression()),
    'Overfit': make_pipeline(PolynomialFeatures(10), LinearRegression())
}

for name, m in models.items():
    y_pred_train = m.fit(X_train, y_train).predict(X_train)
    y_pred_test = m.predict(X_test)
    mse_train = mean_squared_error(y_train, y_pred_train)
    mse_test = mean_squared_error(y_test, y_pred_test)
    status = (
        "Overfitting" if mse_test > mse_train * 1.5
        else "Underfitting" if mse_train > mse_test * 1.5
        else "Optimal Fitting"
    )
    print(f"{name} Model:")
    print("MSE Training:", mse_train)
    print("MSE Testing:", mse_test)
    print("Status Model:", status, "\n")
```
