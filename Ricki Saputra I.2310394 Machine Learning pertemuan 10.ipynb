PCA merupakan metode statistik yang digunakan untuk <b>Mereduksi Dimensi</b> dengan cara mengubah data ke dalam komponen utama (Principal Component) yang memiliki variansi terbesar. PCA digunakan untuk mengatasi masalah dimensi pada data yang memiliki banyak fitur, tanpa kehilangan terlalu banyak informasi penting.

Manfaat PCA:

1. Reduksi Dimensi

2. Visualisasi Data

3. Mengurangi Overfitting (menghapus fitur yang kurang penting

4. Mempercepat Komputasi

###mengambil 5 data dan kemudian menggunakan pca menjadi 3 data

```python
data.info()
```

```python
x = data[['Algebra', 'Calculus1','Calculus2','Statistics']]
y = data[['y']]
```

```python

from sklearn.decomposition import PCA

from sklearn.preprocessing import StandardScaler
```

```python

#Pre proceesing
#Normalisasi Data (Data di-scaling agar memiliki varians yang seragan)

scaler = StandardScaler()

x_scaled =scaler.fit_transform(x)
x_scaled
```

```python
pca = PCA(n_components = 2) #mengurangi 4 fitur menjadi 2 fitur

x_pca = pca.fit_transform(x_scaled)
```

```python
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

x_train, x_test, y_train, y_test = train_test_split(x_pca, y, test_size=0.2, random_state=3)

model = SVC()
model.fit(x_train, y_train.values.ravel())

y_pred_train = model.predict(x_train)
y_pred_test = model.predict(x_test)

print("Accuracy Training =", accuracy_score(y_train, y_pred_train))
print("Accuracy Testing =", accuracy_score(y_test, y_pred_test))
```

```python
x = data[['Algebra', 'Calculus1', 'Calculus2', 'Statistics', 'Probability', 'Measure', 'Functional_analysis']]
y = data[['y']]
```

```python
scaler = StandardScaler()

x_scaled =scaler.fit_transform(x)
x_scaled
```

```python
pca = PCA(n_components = 4) #mengurangi 4 fitur menjadi 2 fitur

x_pca = pca.fit_transform(x_scaled)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_pca, y, test_size=0.2, random_state=3)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC

model_svc = SVC()
model_svc.fit(x_train, y_train)

y_pred_train = model_svc.predict(x_train)
y_pred_test = model_svc.predict(x_test)

akurasi_training = accuracy_score(y_train, y_pred_train)
akurasi_testing = accuracy_score(y_test, y_pred_test)
print(f'Hasil Akurasi Training = {akurasi_training:.2%}')
print(f'Hasil Akurasi Testing = {akurasi_testing:.2%}')
```

###Berdasarkan hasil Pertama, Data yang digunakan untuk traiing hanya 2 fitur saja memberikan akurasi sebesar 61.90% atau (0.6190. Sedangkan data dengan 7 fitur kemudian direduksi menjadi 2 fitur agar informasi data sebelumnya tidak hilang, memberikan hasil akurasi sebesar 63.10%.
