#Decision Tree

Decision Tree adalah metode pembelajaran terawasi (supervised learning) yang digunakan untuk tugas klasifikasi dan regresi. Cara kerjanya adalah dengan membagaj dataset ke dalam subset berdasarkan nilai fitur (feature) atau input, kemudian menciptakan struktur seperti pohon di mana setiap node internal merepresentasikan "uji" pada suatu fitur, setiap cabang merepresentasikan hasil dari uji tersebut, dan setiap node daun merepresentasikan label kelas (untuk klasifikasi) atau nilai (untuk regresi)

##Decision tree classification

Criterion: digunakan untuk metode pemisahan pada node (kotak), menggunakan 'gin' atau 'entropy Entropy. Mengukur ketidakpastian dalam set data, semakin rendah entropy, semakin tinggi nilai informasi yang didapat dari pemisahan gini mengukur seberapa sering elemen dalam set data salah diklasifikasikan jika dipilih secara acak

max depth: mengukur kedalaman maksimal dari pohon untuk mengontrol overfitting

#support vector machine

suport vector machine (svm) adalah algoritma pembelajaran terawasi yang di gunakan untuk tugas klasifikasi dan regresi. algoritma ini bekerja dengan menemukan hyperplane optimal (pada klasifikasi) atau garis pada regresi yang paling baik untuk memisahkan data.

Kernel:

Linear kernel: digunakan ketika data dapat dipisah secara linear. Kernel ini, menghitung data dalam antara dua vektor, dan cocok untuk dataset yang dapat dipisahkan dengan garis lurus.

RBF (Radial Basis Function Kernel): adakah kernel non-linear yang banyak digunakan dalan SVM. Kernel ini yang lenih tinggi membuay daya yang tidak dapat dipisahkan secara linar dakan ruang asli menjad dapat dipisahkan di ruang yang lebih tinggi.

Polynomial Lernel: merupakan kernel nonlinear yang cocok untuk dataset yang memiliki hubungan non linear antara fiturnta. Jadi, SVM ini digunakan untuk memsiahkan data mengguanak kurva polinomial

Sigmoid Kernel: berasal dari fungsi aktivasi sigmoid yang digunakan dalam jaringan saraf tiruan (neutral network). Ini bekerja dengan memetakan titik data ke ruang dimensi yang lebih tinggi dengan mengguanakn fungsi sigmoid.

##suport vector machine classifier

```python
from sklearn.datasets import load_diabetes
from sklearn.svm import SVC

diabetes = load_diabetes()
x = diabetes.data[:, :2]   # Ambil 2 fitur pertama
y = diabetes.target
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Split data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1)

# Model SVR dengan kernel sigmoid
model4 = SVR(kernel='sigmoid')

# Training
model4.fit(x_train, y_train)

# Membuat prediksi
y_pred = model4.predict(x_test)

# Evaluasi model
mse = mean_squared_error(y_test, y_pred)

print(f"Hasil MSE: {mse}")
print("Data Asli:\n", y_test)
print("Data Hasil Prediksi:\n", y_pred)
```

```python

plt.scatter(x_test[:, 0], y_test, color='blue', label='Data Asli')
plt.scatter(x_test[:, 0], y_pred, color='red', label='Data Prediksi')
plt.xlabel('Fitur 1')
plt.ylabel('Fitur 2')
plt.legend()
plt.show()
```
