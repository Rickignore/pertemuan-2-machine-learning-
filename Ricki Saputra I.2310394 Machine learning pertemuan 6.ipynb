-Jelaskan apakah hasil evaluasi tersebut Overfitting, Underfitting, atau Optimal Fitting

```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python

dataset_path = '/kaggle/input/student-performance-factors'
data_files = os.path.join(dataset_path, 'StudentPerformanceFactors.csv')
data = pd.read_csv(data_files)
data.head()
```

```python
x = data[['Hours_Studied', 'Attendance', 'Sleep_Hours', 'Previous_Scores', 'Tutoring_Sessions', 'Physical_Activity']]
y = data[['Exam_Score']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
```

```python

y_pred_train = linear_model.predict(x_train)
y_pred_test = linear_model.predict(x_test)
```

```python
mse_train = mean_squared_error(y_train, y_pred_train)
mse_test = mean_squared_error(y_test, y_pred_test)

status = (
    "Overfitting" if mse_test > mse_train * 1.5
    else "Underfitting" if mse_train > mse_test * 1.5
    else "Optimal Fitting"
)

print("MSE Training:", mse_train)
print("MSE Testing:", mse_test)
print("Status Model:", status)
```

```python
plt.scatter(y_test, y_pred_test, color='blue')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Linear Regression Prediction vs Actual')
plt.show()
```

Hasil evaluasi menunjukkan bahwa MSE Training (6.33) dan MSE Testing (5.06) memiliki nilai yang cukup berdekatan. Ini berarti model mampu mempelajari pola data dengan baik tanpa terlalu menghafal data latih maupun kehilangan kemampuan generalisasi pada data uji. Dengan demikian, model berada dalam kondisi Optimal Fitting — tidak overfitting dan tidak underfitting.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

#Split data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 666)

#Linear Regression
model = LinearRegression()

#Training
model.fit(x_train, y_train)

#Prediksi
y_pred_train = model.predict(x_train) #prediksi data training
y_pred_train = model.predict(x_train) #prediksi data training

#Evaluasi Training
mse_training = mean_squared_error(y_pred_train, y_train)
print('Hasil Evaluasi MSE Training = ', mse_training)

#Evaluasi Testing
mse_testing = mean_squared_error(y_pred_test, y_test)
print('Hasil Evaluasi MSE Testing = ', mse_testing)

#visualilasi
plt.scatter(y_test, y_pred_test, color='blue')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Linear Regression Prediction vs Actual')
plt.show()
```
