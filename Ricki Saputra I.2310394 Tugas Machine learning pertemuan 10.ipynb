##Random Forest untuk 5 fitur

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
```

```python
x = data[['Algebra', 'Calculus1','Calculus2','Statistics','Probability']]
y = data[['y']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
model = RandomForestClassifier()
model.fit(x_train, y_train)

y_pred_train = model.predict(x_train)
y_pred_test = model.predict(x_test)

akurasi_training = accuracy_score(y_train, y_pred_train)
akurasi_testing = accuracy_score(y_test, y_pred_test)

print(f"training = {akurasi_training:.2%}")
print(f"testing = {akurasi_testing:.2%}")
```
